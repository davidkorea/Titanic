````python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, roc_curve, auc, accuracy_score
from sklearn.tree import DecisionTreeClassifier
import pydotplus
````

````python
train = pd.read_csv('./data/train.csv')
test =pd.read_csv('./data/test.csv')
````

````python
train.head()
````

# Data exploration

````python
print('train shape: ', train.shape)
print('test shape: ', test.shape)
````

````python
train.isnull().sum()
````

````python
test.isnull().sum()
````

````python
full_df = pd.concat([train,test], axis=0, ignore_index=True)
full_df.head()
````

````python
full_df.shape
````

````python
full_df.isnull().sum()
````

## 1. Embarked 출발지

````python
full_df.Embarked.value_counts()
````

````python
# because most people are from S
full_df.Embarked = full_df.Embarked.fillna('S')
````

````python
full_df.Embarked.isnull().sum()
````

## 2. Carbin

````python
full_df.drop(['Cabin'],axis=1,inplace=True)
````

## 3. Age

````python
full_df.Age.describe()
````

````python
full_df.Age = full_df.Age.fillna(full_df.Age.mean())
````

````python
full_df.dtypes
````

````python
full_df.columns
````

# Precessing Data

````python
target = train.Survived
target.shape
````

````python
# select usaful feature columns
full_df = full_df[['Age', 'Embarked', 'Fare', 'Parch', 'Pclass','Sex', 'SibSp']]
````

````python
train_df = full_df[:train.shape[0]]
````

````python
train_df.shape
````

## 1. Sex, Embarked  -> 0,1 Labelencode

````python
train_df['Sex'] = train_df['Sex'].map({'male':1,'female':0})
````

````python
train_df['Embarked'] = train_df['Embarked'].map({'C':0,'Q':1,'S':2})
````

-----
**train_df.Embarked vs train_df['Embarked'] 차이점**

```train_df['Embarked'] = train_df['Embarked'].map({'C':0,'Q':1,'S':2})```
-> int64

```train_df.Embarked = train_df.Embarked.map({'C':0,'Q':1,'S':2})```
-> float64

-----

````python
train_df.dtypes
````

````python
train_df.head()
````

````python
train_df.SibSp.value_counts()
````

````python
def get_age(age):
    if age<=12:
        label = 0
    elif age<18:
        label = 1
    elif age<30:
        label = 2
    elif age<60:
        label = 3
    else: label = 4
    return label
train_df['Age'] = train_df['Age'].apply(get_age)
````

````python
train_df.Fare.describe()
````

````python
# 可以按照四分位进行分类
def get_fare(fare):
    if fare<=8:
        label = 0
    elif fare<=14:
        label = 1
    elif fare<=31:
        label = 2
    else: label = 3
    return label

train_df.Fare = train_df.Fare.apply(get_fare)
    
````

````python
train_df.Fare.value_counts()
````

# Modeling

````python
x = train_df
y = train['Survived']
````

````python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/5,random_state=1)
````

````python
print(x_train.shape,x_test.shape)
````

````python
clf = DecisionTreeClassifier(random_state=2018) # all paras by default
````

````python
clf.fit(x_train,y_train)
````

````python
clf.score(x_test,y_test)
````

````python
predictions = clf.predict(x_test)
````

````python
predictions
````

# Evaluation

````python
confusion_matrix(y_test,predictions)
````

````python
accuracy_score(y_test,predictions)
````
